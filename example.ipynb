```python
#导入combo包
import combo
import sympy as sp
```

```python
#定义对易式指标
#left=[['a'],['b']]
left=[['a','b'],['c','d']]
right=[['i','j'],['k','l']]
#right=[['i','j','k'],['l','m','n']]
#right=[['i'],['j']]

operators = combo.bodys(left,right)
```

```python
#计算对易式
operators.commutate()
```

```python
#化简delta项和xi项
operators.applyRule('both')
```

```python
#筛选出想要的项,并对指标规范化
operators.regulate(filterbody=0)
```

```python
#得到初始的未化简的表达式
initial_expr = operators.canon
if len(initial_expr.args) < 100:
    print('Initial expression has {} terms'.format(len(initial_expr.args)))
    combo.jupyterDisplay(initial_expr)
else:
    print('Initial expression has {} terms'.format(len(initial_expr.args)))
```

```python
#对原始表达式进行化简
simplified_expr = combo.simplifyUseBoth(initial_expr)
print('Simplified expression has {} terms'.format(len(simplified_expr.args)))
combo.jupyterDisplay(simplified_expr)
```

```python
combo.jupyterDisplay(combo.uniteSameGAndH(simplified_expr))
```

```python
#筛选出含有不同多体lambda的项
filter_lambda_expr = combo.filterLambdaBody(simplified_expr,filterLambdaBody=1)
combo.jupyterDisplay(filter_lambda_expr)
print(len(filter_lambda_expr.args))
```

```python
#对筛选出来的表达式进行化简，合并具有相同指标的G和H项
combo.jupyterDisplay(combo.uniteSameGAndH(filter_lambda_expr))
```

```python
#进一步筛选
combo.jupyterDisplay(sp.simplify(filter_lambda_expr).args[1])
futher_filter_lambda_expr = combo.filterLambdaBody(sp.simplify(filter_lambda_expr).args[1],filterLambdaBody=2)
remaning_terms = sp.simplify(filter_lambda_expr)/sp.simplify(filter_lambda_expr).args[1]
print(remaning_terms)

combo.jupyterDisplay(futher_filter_lambda_expr)
print(len(futher_filter_lambda_expr.args))

combo.jupyterDisplay(combo.uniteSameGAndH(futher_filter_lambda_expr)*remaning_terms)
```

```python
combo.jupyterDisplay(combo.uniteSameGAndH(futher_filter_lambda_expr)*remaning_terms)
```

```python
combo.jupyterDisplay(sp.simplify(futher_filter_lambda_expr)*remaning_terms)
```

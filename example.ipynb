```python
#导入combo包
import qcombo
```

```python
#定义对易式指标
left=[['a'],['b']]
#left=[['a','b'],['c','d']]
#right=[['i','j'],['k','l']]
#right=[['i','j','k'],['l','m','n']]
right=[['i'],['j']]

operators = qcombo.bodys(left,right)
```

```python
#calculate the commutator
operators.commutate()
#display the Latex of product of left and right operator 
#based on the generalized wick theorem
qcombo.jupyterDisplay(operators.gw)
#display the Latex of commutator of left and right operator 
#based on the generalized wick theorem
qcombo.jupyterDisplay(operators.cmt)
```

```python
#Diagonalize the one-body density matrix
operators.applyRule('both')
#display the commutator after Diagonalizing
qcombo.jupyterDisplay(operators.cmtRule)
```

```python
#Filter the desired terms and normalize the indices
operators.regulate(filterbody=1)
#Filtered terms with Matrix elements
qcombo.jupyterDisplay(operators.filterTerms)
#Contract the delta indices and canonicalize indices
qcombo.jupyterDisplay(operators.canon)
```

```python
#unsimplified expression
initial_expr = operators.canon
#simplify the expression
simplified_expr = qcombo.simplifyUseBoth(initial_expr)
qcombo.jupyterDisplay(simplified_expr)
```

```python
#filter based on the lambda body
filter_lambda_expr = qcombo.filterLambdaBody(simplified_expr,filterLambdaBody=1)
qcombo.jupyterDisplay(filter_lambda_expr)
```

```python
#unite the same terms for simplifying
united_expr = qcombo.uniteSameGAndH(filter_lambda_expr)
qcombo.jupyterDisplay(qcombo.uniteSameGAndH(filter_lambda_expr))
```

```python
from qcombo.output import getEquationLatexStrFromExpr
#trans expresion to latex equation 
latex_str = getEquationLatexStrFromExpr(united_expr,1,1,'R')
print(latex_str)
```

```python
from qcombo.output import getEquationAmcStrFromExpr
#trans expresion to input equation for amc package 
amc_str = getEquationAmcStrFromExpr(united_expr,1,1,'R')
print(amc_str)
```
